```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("boston.csv")
```

```python
df.columns = df.iloc[0]
```

```python
df = df.drop(0)
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
X = df.iloc[:, 0:13]
y = df.iloc[:, 13]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y , train_size=0.80, test_size=0.20, random_state=123)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
```

```python
lr.fit(X_train,Y_train)
```

```python
dt.fit(X_train,Y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = dt.predict(X_test)
```

```python
print("R^2 score for LR",r2_score(Y_test,y_pred1))
print("R^2 score for DT",r2_score(Y_test,y_pred2))
```

```python
from sklearn.ensemble import BaggingRegressor

br = BaggingRegressor(random_state=1)
br.fit(X_train, Y_train)
```

```python
Y_preds = br.predict(X_test)
```

```python
print('Training Coefficient of R^2 : %.3f'%br.score(X_train, Y_train))
print('Test Coefficient of R^2 : %.3f'%br.score(X_test, Y_test))
```
