```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=100, factor=0.1, noise=.1)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
classifier = SVC(kernel="linear")
classifier.fit(X_train, y_train.ravel())
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
zero_one_colourmap = ListedColormap(('blue', 'red'))
def plot_decision_boundary(X, y, clf):
    X_set, y_set = X, y
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                                 stop = X_set[:, 0].max() + 1, 
                                 step = 0.01),
                       np.arange(start = X_set[:, 1].min() - 1, 
                                 stop = X_set[:, 1].max() + 1, 
                                 step = 0.01))
  
    plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), 
                                             X2.ravel()]).T).reshape(X1.shape),
               alpha = 0.75, 
               cmap = zero_one_colourmap)
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = (zero_one_colourmap)(i), label = j)
    plt.title('SVM Decision Boundary')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    return plt.show()
```

```python
plot_decision_boundary(X, y, classifier)
```

```python
def plot_3d_plot(X, y):
    r = np.exp(-(X ** 2).sum(1))
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=100, cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
```

```python
plot_3d_plot(X,y)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_circles

# Generate circular dataset
X, y = make_circles(n_samples=100, factor=.1, noise=.1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

# Define the parameters grid for GridSearchCV
param_grid = {'kernel': ['linear', 'poly', 'rbf', 'sigmoid']}

# Initialize SVC classifier
classifier = SVC()

# Initialize GridSearchCV with SVC classifier and parameter grid
grid_search = GridSearchCV(classifier, param_grid, cv=5, scoring='accuracy')

# Fit the GridSearchCV on training data
grid_search.fit(X_train, y_train)

# Get the best estimator and its parameters
best_estimator = grid_search.best_estimator_
best_params = grid_search.best_params_

# Make predictions on test data using the best estimator
y_pred = best_estimator.predict(X_test)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Best Parameters:", best_params)
print("Accuracy Score:", accuracy)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_circles

# Generate circular dataset
X, y = make_circles(n_samples=100, factor=.1, noise=.1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

# Define the kernels to loop over
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

# Initialize an empty dictionary to store the results
results = {}

# Loop over each kernel
for kernel in kernels:
    # Initialize SVC classifier with current kernel
    classifier = SVC(kernel=kernel)
    
    # Define the parameter grid for GridSearchCV
    param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [0.1, 0.01, 0.001, 0.0001]}
    
    # Initialize GridSearchCV with SVC classifier and parameter grid
    grid_search = GridSearchCV(classifier, param_grid, cv=5, scoring='accuracy')
    
    # Fit GridSearchCV on training data
    grid_search.fit(X_train, y_train)
    
    # Get the best estimator and its parameters
    best_estimator = grid_search.best_estimator_
    best_params = grid_search.best_params_
    
    # Make predictions on test data using the best estimator
    y_pred = best_estimator.predict(X_test)
    
    # Calculate accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    
    # Store the results
    results[kernel] = {'best_params': best_params, 'accuracy': accuracy}

# Print the results
for kernel, result in results.items():
    print("Kernel:", kernel)
    print("Best Parameters:", result['best_params'])
    print("Accuracy Score:", result['accuracy'])
    print()
```
