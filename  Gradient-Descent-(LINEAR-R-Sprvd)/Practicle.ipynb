```python
stores = df.groupby(by='Store')
stores['Weekly_Sales'].mean().sort_values(ascending=False)[:15].plot(kind='bar',color='black')
plt.title('Top 10 stores by Weekly_Sales')
plt.show()
```

```python
def year_wise_sales(year):
    filt = df['Year'] == year
    result = df.loc[filt].groupby(by='Store')
    values = result['Weekly_Sales'].mean()
    return values 

year_2010 = year_wise_sales(2010)
year_2011 = year_wise_sales(2011)
year_2012 = year_wise_sales(2012) 

width = 0.25
x_index = np.arange(len(year_2010))

plt.figure(figsize=(15, 4))
plt.title('Sales of all stores for 3 years')
plt.bar(x_index - width, year_2010, width=width, label='2010')
plt.bar(x_index, year_2011, width=width, label='2011')
plt.bar(x_index + width, year_2012, width=width, label='2012')
plt.xticks(x_index)
plt.legend()
plt.show()
```

```python
plt.title('Total sales per Year')
year_sales = df.groupby(by='Year')
year_sales['Weekly_Sales'].sum().plot(kind='bar', color='black')
plt.show()
```

```python
def season(month):
    if month in [1,2,12]:
        return 'Winter'
    elif month in [3,4,5]:
        return 'Spring'
    elif month in [6,7,8]:
        return 'Summer'
    elif month in [9,10,11]:
        return 'Autumn'
    
df['Season'] = df['Month'].apply(season)
```

```python
season = df.groupby(by='Season')
season['Weekly_Sales'].sum().sort_values(ascending=False).plot(kind='bar', color='black')
plt.title('Season wise Sales')
plt.show()
```

```python
sns.histplot(df['Temperature'], kde=True, color='black')
```

```python
def whether(temp):
    if temp < 32:
        return 'Very Cold'
    elif temp > 32 and temp <50:
        return 'Cold'
    elif temp > 50 and temp< 80:
        return 'Normal'
    elif temp > 80 :
        return 'Hot'
    
df['Weather'] = df['Temperature'].apply(whether)
```

```python
whether = df.groupby(by='Weather')
whether['Weekly_Sales'].mean().sort_values(ascending=False).plot(kind='bar',color='black')
plt.title('Sales for different Whether')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['Season'] = label_encoder.fit_transform(df['Season'])
df['Weather'] = label_encoder.fit_transform(df['Weather'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['Weekly_Sales', 'Date'], axis=1)
y = df['Weekly_Sales']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
model = SGDRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

rmse1 = np.sqrt(mean_squared_error(y_test, y_pred))
rmse1
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)
r2
```
